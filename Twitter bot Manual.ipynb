To compute the frequencies of the words being used in the tweets collected before, we first need to clean the tweets so that website urls,be verbs,etc will not be counted as words. We will also convert all the tweet data into lower case such that for eg: "Apple" and "apple" words are treated equally.

For Cleaning the tweets we used 're' and 'nltk' modules


```python
#now we will perform data cleaning on tweets so that we will get only precise data
#Here we will clean website urls, 'be' verbs, and treat upper case and lower case words equally

cleaned = []
ps = PorterStemmer()
stop_words = set(stopwords.words('english'))
tweets =  tw.Cursor(api.search_tweets, q="Avatar").items(1000)
tweets_text = [tweet.text for tweet in tweets]
for tweet in tweets_text:
    cleaned_data = re.sub("(@[A-Za-z0-9]+)|([^0-9A-Za-z \t])|[(\w+:\/\/\S+)|[0-9]]",' ',tweet)
    cleaned_data = cleaned_data.lower()
    cleaned_data = cleaned_data.split()
    cleaned_data = [x for x in cleaned_data if not x in stop_words]
    cleaned_data = " ".join(cleaned_data)
    cleaned.append(cleaned_data)
```

Now Lets see all the words according to their frequency

```python
#Now the clened data is split into words

words = []
for i in range(0,len(tweets_text)):
    words.append(cleaned[i].split())

#Shows all the frequencies of the words being used in the tweets

word = itertools.chain(*words)
frequent_words = collections.Counter(word)
pd.DataFrame(frequent_words.most_common())
```

Now Lets see the Top 30 words which are used in the Tweets

```python
pd.DataFrame(frequent_words.most_common(30))
```

Top 10 hashtags, top 10 user mentions that are the most popular in the collection of tweets.

```python
tweets =  tw.Cursor(api.search_tweets, q="Avatar").items(1000)

hashtags_dict = {}
for tweet in tweets:
    hashtags = tweet.entities.get('hashtags')
    for hashtag in hashtags:
        if hashtag['text'] in hashtags_dict.keys():
            hashtags_dict[hashtag['text']] += 1
        else:
            hashtags_dict[hashtag['text']] = 1
top_10_hashtags = sorted(hashtags_dict, key=hashtags_dict.get, reverse=True)
top_10_hashtags = top_10_hashtags[:10]
for i in range (0,10):
    print(top_10_hashtags[i])
```

```python
tweets =  tw.Cursor(api.search_tweets, q="Avatar").items(1000)

user_mentions_dict = {}
for tweet in tweets:
    user_mentions = tweet.entities.get('user_mentions')
    for mention in user_mentions:
        if mention['screen_name'] in user_mentions_dict.keys():
            user_mentions_dict[mention['screen_name']] += 1
        else:
            user_mentions_dict[mention['screen_name']] = 1
top_10_user_mentions = sorted(user_mentions_dict, key=user_mentions_dict.get, reverse=True)
top_10_user_mentions = top_10_user_mentions[:10]
for i in range(0,10):
    print(top_10_user_mentions[i])
```
